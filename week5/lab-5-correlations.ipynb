# 3. Scatterplots and Trend Lines
Please refer to [Lab 3 -> Scatterplot](https://ctools.umich.edu/access/content/group/09aeadb1-5ac1-4433-9741-033966380c8a/Week%204%20-%20Basic%20Vis/Lab%203/lab-3-vis-with-solutions.html#3.1-Scatter-Plot) for a review on how to create basic scatterplots. We will explore some variants of scatterplots here.

```python
# Scatterplots with discrete values.
# Let's create a dimension with discrete values by rounding up 
# petal_ratio to the nearest integer.
df['petal_length_rounded'] = df['petal_length'].round()
sns.lmplot(x='petal_length_rounded', y='petal_width', 
           data=df, fit_reg=False, hue='species');
```

```python
# In the above plot, since the X dimension is discrete, many data 
# points are on top of another, and it is hard to distinguish them.
# To solve this issue, we can add some "jittering" (small noise) on
# the x-axis, so as to improve the readability of the plot without 
# distorting the data too much.
sns.lmplot(x='petal_length_rounded', y='petal_width', 
           data=df, fit_reg=False, hue='species', x_jitter=.1);
```

```python
# See the difference?
# We can also add a trend (regression) line to the plot.
sns.lmplot(x='petal_length_rounded', y='petal_width', 
           data=df, fit_reg=True, x_jitter=.1);
```

```python
# We can show aggregate statistics for each value of the discrete variable.
# We see means and confidence intervals for the unique values of the X-dim 
# variable.
sns.lmplot(x='petal_length_rounded', y='petal_width',
           data=df, x_estimator=np.mean);
```

```python
# We may do binning, too. This plot places a continuous X-dim variable 
# into discrete bins. Notice that I changed back to "x=petal_length".
sns.lmplot(x='petal_length', y='petal_width', data=df, x_bins=8);
```

```python
# Compare the above plot with the original scatterplot.
sns.lmplot(x='petal_length', y='petal_width', data=df);
```

```python
# We can add a 2nd-order regression line instead of a linear regressoin line,
# although it is not necessary for this particular plot.
# Notice how the trend line curves a bit?
sns.lmplot(x='petal_length', y='petal_width', data=df, order=2);
plt.xlim(0, 8);
plt.ylim(-0.5, 3.0);
```

```python
# Finally, we can create scatter plots with marginal histograms.
sns.jointplot(x='petal_length', y='petal_width',
              data=df, kind='reg');
```

# 5. Correlation Metrics

__Reference__:
A good article explaining the differences between three types of correlations: [Correlation (Pearson, Kendall, Spearman)](http://www.statisticssolutions.com/correlation-pearson-kendall-spearman/).

```python
# Calculate the correlation between sepal_length and each 
# of the three other measurements, using two different 
# correlation metrics. Visualize the result using a barplot.
corr_metrics = ['pearson', 'spearman']
other_measurements = ['sepal_width', 'petal_length', 
                      'petal_width']
cols = {m: [df['sepal_length'].corr(df[x], method=m) 
            for x in other_measurements] 
        for m in corr_metrics}
df_corr = pd.DataFrame(cols, index=other_measurements,
                       columns=corr_metrics)
sns.set(style='whitegrid')
df_corr.plot(kind='barh', 
             title='Correlation with sepal_length');
```

# 6. Regression Diagnostics
We learned how to use Cook's distance and DFFITS to identify outliers. Here, we demonstrate various toolkits to identify influential points in a regression model. We will use Anscombe's quartet for the demonstrations.

```python
# This is the Anscombe's quartet
# Source - Wikipedia: https://en.wikipedia.org/wiki/Anscombe%27s_quartet
from io import StringIO

TESTDATA=StringIO("""X1,Y1,X2,Y2,X3,Y3,X4,Y4
10,8.04,10,9.14,10,7.46,8,6.58
8,6.95,8,8.14,8,6.77,8,5.76
13,7.58,13,8.74,13,12.74,8,7.71
9,8.81,9,8.77,9,7.11,8,8.84
11,8.33,11,9.26,11,7.81,8,8.47
14,9.96,14,8.1,14,8.84,8,7.04
6,7.24,6,6.13,6,6.08,8,5.25
4,4.26,4,3.1,4,5.39,19,12.5
12,10.84,12,9.13,12,8.15,8,5.56
7,4.82,7,7.26,7,6.42,8,7.91
5,5.68,5,4.74,5,5.73,8,6.89""")

df_ans = pd.DataFrame.from_csv(TESTDATA, index_col=None)

fig, axs = plt.subplots(2, 2, figsize=(8, 7))
for i, ax in enumerate(axs.flat):
    j = i + 1
    ax.scatter(df_ans['X%d'%j], df_ans['Y%d'%j])
    ax.set_xlabel('X%d'%j)
    ax.set_ylabel('Y%d'%j)
    ax.grid(True)
```

## 6.1 Build a Linear Regression Model

```python
# Let's build a simple linear regression model with Y3 as the 
# dependent variable and X3 as the independent variable.
model1 = smf.ols('Y3 ~ X3', data=df_ans).fit()

# Plot the regression line together with the data
xs = np.linspace(3, 15, 2)
ys_predicted = model1.predict({'X3': xs})
plt.scatter(df_ans.X3, df_ans.Y3)
plt.grid(True)
plt.plot(xs, ys_predicted, 'r');
```

```python
# Compare the above plot with seaborn's linear regressin plot:
sns.lmplot(x='X3', y='Y3', data=df_ans);
```

The `statsmodels` package automatically performs multiple influence tests when you fit a regression model. They can be accesssed by using the `get_influence()` function.

## 6.2 Perform Influence Tests

```python
# Perform influence test
infl1 = model1.get_influence()
infl1.summary_frame().head()
```

### Explanations of the influnece statistics

To understand the above numbers, here are some more important terminologies:
- __Influential point__: synonym to __influential observation__ or __unusual observation__. An influential ponit is a data point whose removal from the dataset would result in a large change in the fit. An influential point may or may not be an outlier and may or may not have large leverage.
- __Outlier__: a data point that has an unusually large residual (Y-direction distance to the fitted line). An outlier may or may not be a leverage point.
- __Leverage point__: a data point that is located extremeley in X-dimension. This property does not depend on Y. A leverage point may or may not be an outlier.

* * *

In the above table, notice there is a row for each observation in the sample. The columns are diagnostic statistics. Almost all of them can be used to identify outliers.
- __DFBETA (`dfb_*`)__: measures the difference in each parameter estimate with and without a given observation point. Therefore, naturally, There is a dfb\_\* statistic for each observation for each parameter.
- __Cook's distance (`cooks_d`)__: measures the distance between the fitted values (Y) calculated with and without a given observation point.
- __DFFTIS (`dffits`)__: same as Cook's distance, except the distance is measured by the approximate number of standard deviations.
  - __`dffits_internal`__ measures the influence of an observation based on internally studentized residuals, which simply means the observation point in question is included when computing the standard deviation for the purpose of normalization.
- __Leverage (`hat_diag`)__: measures the distance from an observation's X-value to the average of the X-values of all observations in the sample. It does not depend on Y-dim values. Useful for identifying observations that have unusual predictor values compared to the remaining data. 
  - Note that the leverage of the $i$-th observation point is defined as $h_{ii} = (H)_{ii}$, i.e., the $i$-th diagonal element of the __hat matrix__ $H=X(X'X)^{-1}X'$, which is also called influence matrix or projection matrix.
- __Residual (`standard_resid`)__: The distance between the fitted value ($\hat{Y}$) and the actual observation value ($Y$).
- __Studentized Residuals__: Quotient resulting from the division of a residual by an estimate of its standard deviation, which is deteremined by leverage.
  - __Internally Studentized Residuals__: The observation point in question is included when computing the standard deviation.
  - __Externally Studentized Residuals (`student_resid`)__: The observation in question is not included when computing the standard deviation. DFFITS equals Externally Studentized Residuals times $\sqrt{h_{ii}/(1-h_{ii})}$, where $h_{ii}$ is the leverage for the observation point.
    
* * * 

In practice, the following __cut-off thresholds__ may be used to identify outliers or leverage points. Let us define n as the number of observations, and p as the number of parameters.
- __DFBETA__: 2/sqrt(n). Observation points whose absolute DFBETA values greater than this threshold are considered influential points. If the threshold is greater than 1, use 1.
- __Cook's distance__: 4/n, or simply, 1. Observation points whose Cook's distance is greater than this threshold are considered as influential points.
- __DFFITS__: 2 * sqrt(p/n). Observation points whose absolute DFFITS values are greater than  this threshold are considered influential points.
- __Leverage__: 3p/n. Observations whose leverage is greater than this threshold are considered leverage points. If the threshold is smaller than 0.99, use 0.99. Note that leverage is always between [0, 1].

## 6.3 Produce Influence Plots
### 6.3.1 Residual-Leverage Plot
This is a plot that is helpful for identifying
- __leverage points__ on the extremes of the X-axis
- __outliers__ on the extremes of the Y-axis.
- __influence points__ by the size of the points (Cook's distance or DFFITS).

Notice that the influence points are labeled with their indexes.

```python
# Influence plot of Studentized Residual vs. Leverage.
fig, ax = plt.subplots(figsize=(10,7))
fig = sm.graphics.influence_plot(model1, ax=ax, criterion='dffits')
ax.set_xlim(0, 0.35)
```

### 6.3.2 Leverage vs. Squared-Residual plot
Similar to the above plot, except the residuals are squared, and X and Y axises are swapped, and the nodes are not sized by influence.

```python
fig, ax = plt.subplots()
fig = sm.graphics.plot_leverage_resid2(model1, ax=ax)
```

### 6.3.3 Regression Fit Plot
Plot fit against one regressor. This creates one graph with the scatterplot of observed values compared to fitted values.

```python
fig, ax = plt.subplots()
fig = sm.graphics.plot_fit(model1, 'X3', ax=ax)
```

```python
# This one creates even more plots. You are welcome to research their meanings on your own.
fig = plt.figure(figsize=(10,7))
fig = sm.graphics.plot_regress_exog(model1, 'X3', fig=fig)
```
