```python
plt.figure(figsize=(12,8))
sns.boxplot(x='measure', y='value', data=df_stacked, hue='species')
sns.stripplot(x='measure', y='value', data=df_stacked, jitter=True, size=2, hue='species')
```

```python
df_setosa = df_stacked[(df_stacked.species=='setosa')]
measures = ['sepal_length', 'sepal_width']
df_setosa = df_setosa[df_setosa.measure.isin(measures)]
df_setosa.head()
sns.boxplot(x='measure', y='value', data=df_setosa)
sns.stripplot(x='measure', y='value', data=df_setosa, jitter=True, size=2)
plt.title("Distribution of sepal length and sepal width for Setosa")
```

```python
corr_metrics = ['pearson', 'spearman']
other_measurements = ['sepal_width', 'petal_length', 'petal_width']
cols = {m: [df['sepal_length'].corr(df[x], method=m) for x in other_measurements] for m in corr_metrics}
df_corr = pd.DataFrame(cols, index=other_measurements, columns=corr_metrics)
df_corr
sns.set(style='whitegrid')
df_corr.plot(kind='barh', title='Correlation with sepal_length')
```

```python
from io import StringIO

TESTDATA=StringIO("""X1,Y1,X2,Y2,X3,Y3,X4,Y4
10,8.04,10,9.14,10,7.46,8,6.58
8,6.95,8,8.14,8,6.77,8,5.76
13,7.58,13,8.74,13,12.74,8,7.71
9,8.81,9,8.77,9,7.11,8,8.84
11,8.33,11,9.26,11,7.81,8,8.47
14,9.96,14,8.1,14,8.84,8,7.04
6,7.24,6,6.13,6,6.08,8,5.25
4,4.26,4,3.1,4,5.39,19,12.5
12,10.84,12,9.13,12,8.15,8,5.56
7,4.82,7,7.26,7,6.42,8,7.91
5,5.68,5,4.74,5,5.73,8,6.89""")

df_ans = pd.DataFrame.from_csv(TESTDATA, index_col=None)
fig, axs = plt.subplots(2, 2, figsize=(8,7))
for i,ax in enumerate(axs.flat):
    j = i+1
    ax.scatter(df_ans['X%d'%j], df_ans['Y%d'%j])
    ax.set_xlabel('X%d'%j)
    ax.set_ylabel('Y%d'%j)
    ax.grid(True)
```

```python
model1 = smf.ols('Y3 ~ X3', data=df_ans).fit()

xs = np.linspace(3,15,2)
ys_predicted = model1.predict({'X3': xs})
plt.scatter(df_ans.X3, df_ans.Y3)
plt.grid(True)
plt.plot(xs, ys_predicted, 'r')
```

```python
sns.lmplot(x='X3', y='Y3', data=df_ans)
```

```python
infl1 = model1.get_influence()
infl1.summary_frame().head()
```

```python
fig, ax = plt.subplots(figsize=(10,7))
fig = sm.graphics.influence_plot(model1, ax=ax, criterion='dffits')
```

```python
fig, ax = plt.subplots(figsize=(10,7))
fig = sm.graphics.plot_leverage_resid2(model1, ax=ax)
```

```python
fig, ax = plt.subplots(figsize=(10,7))
fig = sm.graphics.plot_fit(model1, 'X3', ax=ax)
```

```python
fig = plt.figure(figsize=(10,7))
fig = sm.graphics.plot_regress_exog(model1, 'X3', fig=fig)
```

```python
df_virginica = df[df.species=='virginica']
model2 = smf.ols('petal_length ~ petal_width', data=df_virginica).fit()
plt.scatter(df_virginica['petal_width'], df_virginica['petal_length'])

xs = np.linspace(1,7,2)
ys_predicted = model2.predict({'petal_width': xs})
plt.plot(xs, ys_predicted, 'r')
plt.xlim(1.2,2.6)
plt.ylim(4,7.5)
```
