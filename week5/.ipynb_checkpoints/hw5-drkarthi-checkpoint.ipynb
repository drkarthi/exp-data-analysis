## f) Create a scatterplot matrix (SPLOM) with the variables of Budget, Gross, ProfitMargin, RunningTime, and RottenTomatoesRating.
* Limit to drama films only.
* Drop those films whose ProfitMargin is smaller than -1.
* Describe which variable exhibits a positive correlation with RunningTime by reading the SPLOM.

```python
df_splom = df[df.ProfitMargin >= -1]
sns.pairplot(df_splom[['Budget', 'Gross', 'ProfitMargin', 'RunningTime', 'RottenTomatoesRating']], kind='reg')
```

### Budget clearly exhibits a positive correlation with running time

## (g) Calculate the correlation between RunningTime and each of the three variables: Budget, Gross, and ProfitMargin. Use a barchart to present the result.
* Use three different correlation metrics: Pearson, Spearman, and Kendall.
* Use the same filtered data as in 2(f).
* Describe what you see in the visualization.

```python
corr_metrics = ['pearson', 'spearman', 'kendall']
corr_variables = ['Budget', 'Gross', 'ProfitMargin']
cols = {m: [df_splom['RunningTime'].corr(df_splom[x], method=m) for x in corr_variables] for m in corr_metrics}
df_corr = pd.DataFrame(cols, index=corr_variables)
df_corr.plot(kind='barh')
plt.title('Correlation with Running Time')
```

## (h) Create a scatterplot with linear fit for Budget vs. RunningTime. Use the same data as in 2(f).

```python
sns.lmplot(x='RunningTime', y='Budget', data=df_splom)
```

## (i) Build a linear regression model of Budget from RunningTime, using the same data as in 2(f). Display the summary of the model.

```python
model1 = smf.ols('Budget ~ RunningTime', data=df_splom).fit()
model1.summary()
```

## j) Below is Trend Lines Model description generated using Tableau with the same data.

    * (j-1): residual degrees of freedom (DF)
    * (j-2): SSE (sum squared error)
    * (j-3): MSE (mean squared error)
    * (j-4): R-Sqaured
    * (j-5): p-value (significance)

## (k) Use a diagnostic statistic of your choice (e.g., Cook's distance) to identify outliers in the regression model. Filter the outliers, and build a new regression model.
* Compare the R-squared values.
* Create a scatterplot with the regression lines of both the old and the new models.

```python
infl1 = model1.get_influence()
cooks_distance = infl1.cooks_distance[0]
cooks_threshold = 4 / len(cooks_distance)
cooks_threshold

df_infl_points = df_splom[(cooks_distance > cooks_threshold)]
df_infl_points.shape

outlier_criterion = (cooks_distance > cooks_threshold)
df_filtered = df_splom[~ outlier_criterion]
model2 = smf.ols('Budget ~ RunningTime', data=df_filtered).fit()
model2.summary()
```

### The R-squared value is greater for this model (0.438 as compared to 0.408)

```python
xs = np.linspace(80, 180, 2)
ys1 = model1.predict({'RunningTime': xs})
ys2 = model2.predict({'RunningTime': xs})

plt.scatter(df_splom['RunningTime'], df_splom['Budget'])
plt.scatter(df_infl_points['RunningTime'], df_infl_points['Budget'], color='r', s=80, marker='v')

plt.plot(xs, ys1, label="Before outlier removal")
plt.plot(xs, ys2, 'r--', label="After outlier removal")
plt.legend(loc="upper left")
```

## 3(a) Based on your recommended creative type and production budget in Problem 2, perform additional analysis and recommend the following to George regarding the new movie:
* recommend a director
* recommend a distributor


```python
df_recommended = df[(df.CreativeType=='Contemporary Fiction') & (df.Budget < 50000000)]
df_recommended.head()

df_pivot_table = pd.pivot_table(df_recommended, index=['Director'], values=['ProfitMargin'], aggfunc=[np.mean,len])
df_pivot_table[ (df_pivot_table['mean'].ProfitMargin > 0.6) & (df_pivot_table['len'].ProfitMargin > 1)]
```

### Recommended director - Clint Eastwood

```python
df_pivot_table_distri = pd.pivot_table(df_recommended, index=['Distributor'], values=['ProfitMargin'], aggfunc=[np.mean,len])
df_pivot_table_distri[(df_pivot_table_distri['mean'].ProfitMargin > 0.8) & (df_pivot_table_distri['len'].ProfitMargin > 1)]
```
