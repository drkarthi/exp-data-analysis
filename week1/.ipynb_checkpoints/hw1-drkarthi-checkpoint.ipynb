```python
# import libraries
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
# read the data
df_1995 = pd.read_csv('names/yob1995.txt')
years = range(1880,2015)
pieces = []
for year in years:
    path = 'names/yob%d.txt'%year
    frame = pd.read_csv(path)
    frame['year'] = year
    pieces.append(frame)
df_all = pd.concat(pieces, ignore_index=True)
```

```python
# 4c)
df_1995_girls = df_1995[(df_1995.gender=='F')]
df_1995_girls.sort('birth_count', ascending=False).head(10)
```

```python
# 5b)
df_all.groupby(['year', 'gender']).sum()
```

```python
# 5c) Which year has the maximum total #births for boys?
df_counts_by_year = df_all.groupby('year').sum()
year_with_max_count = np.argmax(df_counts_by_year.birth_count)
df_counts_by_year.ix[[year_with_max_count]]
```

```python
# 5d) Which year has the maximum total #births for boys?
df_males = df_all[(df_all.gender=='M')]
df_males_by_year = df_males.groupby('year').sum()
year_with_max_boys = np.argmax(df_males_by_year['birth_count'])
df_males_by_year.ix[[year_with_max_boys]]
```

```python
# 5e) Which year has the biggest difference between #boys and #girls (i.e., abs(#boys-#girls))?
df_pivot = df_all.pivot_table(index="year", columns="gender", aggfunc="sum")
year_with_max_diff = np.argmax( abs(df_pivot['birth_count']['M'] - df_pivot['birth_count']['F']) )
df_pivot.ix[[year_with_max_diff]]
```

```python
# 5f) What are the 5 most popular boys' names in the 1990s (1990-1999)?
df_males_90s = df_all[(df_all.gender=='M') & (df_all.year>=1990) & (df_all.year<=1999)]
df_male_name_counts_90s = df_males_90s.groupby('name')[['birth_count']].sum()
df_male_name_counts_90s.sort('birth_count', ascending=False).head(5)
```

```python
# 5g) What are the 5 most popular girls' names in the 20th century (1901-2000)?
df_girls_20th_century = df_all[(df_all.gender=='F') & (df_all.year>1900) & (df_all.year<=2000)]
df_girls_count_by_name = df_girls_20th_century.groupby('name')[['birth_count']].sum()
df_girls_count_by_name.sort('birth_count', ascending=False).head(5)
```

```python
# 5h) Define the gender neutrality of a name as min(#boys, #girls).
# What are the 15 names with the highest gender neutrality in the 1990s?
df_90s = df_all[(df_all.year>=1990) & (df_all.year<=1999)]
df_pivot = df_90s.pivot_table(index='name', columns='gender', fill_value=0, aggfunc="sum")[['birth_count']]
gender_nuetrality = df_pivot['birth_count'].min(axis=1).to_frame(name="gender_nuetrality")
gender_nuetrality.sort('gender_nuetrality', ascending=False).head(15)
```

```python
# 6a) Plot the total #births by gender and year, as below.
df_males = df_all[(df_all.gender=='M')]
df_females = df_all[(df_all.gender=='F')]
df_male_counts_by_year = df_males.groupby('year').sum()
df_male_counts_by_year.columns = ['M']
df_female_counts_by_year = df_females.groupby('year').sum()
df_female_counts_by_year.columns = ['F']
axs = df_male_counts_by_year.plot()
df_female_counts_by_year.plot(title="Total births by gender and year",ax=axs)
# pivot table is a better solution
```

```python
# 6b) Plot the proportion of the name "Mike" among all male names by year, as below.
df_mike = df_males[(df_males.name=="Mike")]
df_mike_counts_by_year = df_mike.groupby('year').sum()
df_male_counts_by_year = df_males.groupby('year').sum()
df_prop_mike_by_year = df_mike_counts_by_year/df_male_counts_by_year
df_prop_mike_by_year.columns = ['prop']
df_prop_mike_by_year.plot(title="Proportion of \"Mike\" among boy names")
```
