```python
import pandas as pd
years = range(1880,2015)
pieces = []
for year in years:
    path = "names/yob%d.txt"%year
    frame = pd.read_csv(path)
    # add a new column
    frame['year'] = year   
    pieces.append(frame)
df_names = pd.concat(pieces, ignore_index=True)    
```

```python
df_mike = df_names[(df_names.name=='Mike') & (df_names.gender=='M')]
```

```python
%matplotlib inline
df_mike.plot('year','birth_count',title="Mike")
```

```python
df_john = df_names[(df_names.name=='John') & (df_names.gender=='F')]
```

```python
df_john.plot('year', 'birth_count', title="John")
```

```python
def plot_trend(name, gender):
    df_n = df_names[(df_names.name==name) & (df_names.gender==gender)]
    df_n.plot('year','birth_count',title=name)
```

```python
df_mike = plot_trend("Mike", "M")
```
